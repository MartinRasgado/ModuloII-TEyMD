## I. Probabilidad Cl√°sica

La **probabilidad cl√°sica** es una de las formas m√°s antiguas y fundamentales para calcular la probabilidad de que ocurra un evento. Se basa en el supuesto de que todos los resultados posibles de un experimento son **igualmente probables**, es decir, no hay ninguna raz√≥n para pensar que un resultado sea m√°s probable que otro.

La f√≥rmula general es:
$
P(A) = \frac{\text{N√∫mero de casos favorables al evento A}}{\text{N√∫mero total de casos posibles}}
$

Esta definici√≥n se aplica cuando el experimento tiene un n√∫mero **finito y conocido** de posibles resultados, y cada uno de esos resultados tiene la **misma probabilidad** de ocurrir.

---

### üß† ¬øCu√°ndo se aplica la probabilidad cl√°sica?

La probabilidad cl√°sica se usa en situaciones controladas y simples, donde:
- Se puede listar f√°cilmente todos los resultados posibles.
- Se sabe con certeza que todos los resultados tienen la misma probabilidad.
- Los eventos no dependen del azar complejo ni de la frecuencia observada.

---

### üé≤ Ejemplo cl√°sico:

Al lanzar un dado de seis caras, cada cara tiene la misma probabilidad de salir. Si queremos calcular la probabilidad de que salga un n√∫mero mayor que 4, observamos que:

- Casos favorables: {5, 6} ‚Üí hay 2.
- Casos posibles: {1, 2, 3, 4, 5, 6} ‚Üí hay 6.

Entonces:
$
P(\text{n√∫mero > 4}) = \frac{2}{6} = \frac{1}{3} \approx 0.333
$

---

### üìå Limitaciones

La probabilidad cl√°sica **no se puede aplicar** en todos los contextos. No es adecuada cuando:
- No todos los resultados son igualmente probables.
- Es dif√≠cil o imposible conocer todos los resultados posibles.
- Se requiere evidencia emp√≠rica para estimar probabilidades (en esos casos se usa la probabilidad frecuencial o bayesiana).

---

### ‚úÖ Conclusi√≥n

La probabilidad cl√°sica es √∫til para problemas simples y controlados, como juegos de azar, donde se pueden enumerar todos los resultados posibles y se conoce que todos tienen la misma probabilidad. Es una herramienta fundamental para introducir los conceptos b√°sicos de la teor√≠a de la probabilidad.

## II. Probabilidad Geom√©trica

La **probabilidad geom√©trica** es un enfoque que se aplica cuando los eventos se relacionan con medidas continuas como **longitud, √°rea o volumen**. A diferencia de la probabilidad cl√°sica, que se basa en conteo de casos favorables y posibles, la probabilidad geom√©trica se basa en la **proporci√≥n de una medida** dentro de un espacio.

---

### üßÆ Definici√≥n formal:

Si un evento \( A \) ocurre dentro de un espacio continuo y la probabilidad se distribuye de manera uniforme, entonces:
$
P(A) = \frac{\text{Medida del subconjunto favorable (longitud, √°rea o volumen)}}{\text{Medida total del espacio muestral}}
$

---

### üß† ¬øCu√°ndo se usa?

La probabilidad geom√©trica se aplica cuando:

- El n√∫mero de resultados posibles es **infinito** (por ejemplo, infinitos puntos en un segmento).
- Se puede calcular una proporci√≥n basada en geometr√≠a.
- No se puede contar directamente los casos favorables, pero s√≠ medirlos.

---

### üìê Ejemplo:

Supongamos que elegimos al azar un punto sobre un segmento de longitud 10 unidades. ¬øCu√°l es la probabilidad de que ese punto est√© dentro de los primeros 3 unidades del segmento?

- Longitud favorable: 3
- Longitud total: 10

Entonces:
$
P(\text{punto en [0, 3]}) = \frac{3}{10} = 0.3
$

Este resultado indica que hay un **30% de probabilidad** de que el punto caiga dentro del intervalo [0, 3].

---

### üåç Ejemplo con √°rea:

Si un dardo se lanza al azar hacia un blanco cuadrado de 1 m¬≤, y en el centro hay un c√≠rculo de radio $( r $), la probabilidad de que el dardo caiga dentro del c√≠rculo es:
$
P = \frac{\pi r^2}{1}
$

Es decir, se compara el √°rea del c√≠rculo (evento favorable) con el √°rea total del cuadrado (espacio muestral).

---

### üìå Observaci√≥n

Este tipo de probabilidad **no depende del conteo**, sino de **relaciones entre medidas geom√©tricas**. Tambi√©n se asume que **todas las posiciones dentro del espacio tienen la misma probabilidad** de ser elegidas (uniformidad).

---

### ‚úÖ Conclusi√≥n

La probabilidad geom√©trica es √∫til en situaciones continuas donde no se puede contar, pero s√≠ medir. Este enfoque conecta la teor√≠a de la probabilidad con la geometr√≠a, y es com√∫n en problemas f√≠sicos, de localizaci√≥n o de tiro aleatorio en el espacio.

## III. Probabilidad Frecuentista

La **probabilidad frecuentista** se basa en la observaci√≥n de la frecuencia con la que ocurre un evento en un n√∫mero elevado de repeticiones de un experimento. A diferencia de la probabilidad cl√°sica, que parte de una suposici√≥n te√≥rica de casos igualmente probables, la frecuentista depende de **datos emp√≠ricos** obtenidos mediante la repetici√≥n del experimento.

---

### üìê Definici√≥n formal:

La probabilidad de un evento \( A \) se define como el **l√≠mite** de la frecuencia relativa de ocurrencia de $( A $) cuando el n√∫mero de ensayos $( n $) tiende a infinito:
$
P(A) = \lim_{n \to \infty} \frac{f_n(A)}{n}
$

Donde:
- $( f_n(A) )$ es el n√∫mero de veces que ocurre el evento \( A \) en \( n \) repeticiones del experimento.
- $( \frac{f_n(A)}{n} )$ es la **frecuencia relativa** del evento.

---

### üß† En qu√© se basa

- Esta interpretaci√≥n **no asume** que todos los resultados son igualmente probables.
- Se fundamenta en la **repetici√≥n de experimentos** en condiciones similares.
- Cuanto mayor sea el n√∫mero de observaciones, m√°s confiable ser√° la estimaci√≥n de la probabilidad.

---

### üé≤ Ejemplo:

Si lanzamos una moneda 1000 veces y observamos que sali√≥ cara en 498 ocasiones, la probabilidad frecuentista de que salga cara se estima como:
$
P(\text{cara}) \approx \frac{498}{1000} = 0.498
$

A medida que el n√∫mero de lanzamientos aumenta, esta proporci√≥n tiende a estabilizarse cerca de su valor te√≥rico (en este caso, 0.5 si la moneda es justa).

---

### üîÅ Importancia

Este enfoque es ampliamente utilizado en:
- **Estad√≠stica emp√≠rica**
- **Experimentos cient√≠ficos**
- **Simulaciones**
- **Control de calidad y pruebas industriales**

---

### üìå Observaci√≥n

La probabilidad frecuentista **no se puede aplicar a eventos √∫nicos o no repetibles** (como predecir si llover√° ma√±ana o si una persona es culpable). Para esos casos, otras interpretaciones como la bayesiana pueden ser m√°s adecuadas.

---

### ‚úÖ Conclusi√≥n

La probabilidad frecuentista interpreta la probabilidad como una **frecuencia observada**. Es √∫til cuando se puede repetir un experimento muchas veces bajo las mismas condiciones. Se enfoca en la **evidencia emp√≠rica** y es la base de muchas t√©cnicas estad√≠sticas modernas.

## IV. Probabilidad Condicional e Independencia

### üéØ Probabilidad Condicional

La **probabilidad condicional** es la probabilidad de que ocurra un evento $( A $), dado que ya ha ocurrido otro evento $( B $). Se denota como $( P(A \mid B) $), y se define como:
$
P(A \mid B) = \frac{P(A \cap B)}{P(B)} \quad \text{con } P(B) > 0
$

Donde:
- \( P(A \cap B) \) es la probabilidad de que ocurran ambos eventos.
- \( P(B) \) es la probabilidad del evento que ya ocurri√≥.

Esta definici√≥n nos permite **actualizar** nuestra probabilidad sabiendo que cierta informaci√≥n ya es verdadera.

---

### üìå Ejemplo:

Supongamos que en una bolsa hay 5 bolas rojas y 3 azules. Se extrae una bola al azar **sin reemplazo**, y luego una segunda.

Queremos saber la probabilidad de que la **segunda bola sea azul, dado que la primera fue roja**.

- Total inicial: 8 bolas.
- Si la primera fue roja, quedan 7 bolas: 4 rojas y 3 azules.

Entonces:
$
P(\text{azul} \mid \text{roja}) = \frac{3}{7}
$

---

### üîó Independencia de eventos

Dos eventos \( A \) y \( B \) son **independientes** si el hecho de que uno ocurra **no afecta** la probabilidad de que ocurra el otro. Matem√°ticamente, se cumple:

$
P(A \mid B) = P(A)
\quad \text{o equivalentemente} \quad
P(A \cap B) = P(A) \cdot P(B)
$

Esto significa que el conocimiento de uno de los eventos no aporta ninguna informaci√≥n sobre el otro.

---

### üß† Ejemplo de independencia:

Al lanzar dos monedas, sea:
- $( A $): "la primera moneda cae en cara".
- $( B $): "la segunda moneda cae en cara".

Ambos eventos son independientes, ya que el resultado de una moneda **no influye** en el de la otra. Por lo tanto:
$
P(A \cap B) = P(A) \cdot P(B) = 0.5 \cdot 0.5 = 0.25
$

---

### üö´ No confundir:

- Si $( P(A \mid B) \ne P(A) $), los eventos son **dependientes**.
- Todos los eventos independientes cumplen la f√≥rmula del producto:  
  $( P(A \cap B) = P(A) \cdot P(B) $)

---

### ‚úÖ Conclusi√≥n

La probabilidad condicional permite **ajustar** nuestras creencias a partir de nueva informaci√≥n.  
El concepto de independencia es clave en probabilidad, pues simplifica muchos c√°lculos al asumir que ciertos eventos no se afectan mutuamente.

---
## V. Teoremas de la Probabilidad Total y de Bayes

Estos dos teoremas permiten **calcular probabilidades complejas** a partir de otras m√°s simples. Son fundamentales cuando un evento puede depender de varios escenarios previos o causas posibles.

---

### üìê Teorema de la Probabilidad Total

Este teorema permite calcular la probabilidad de un evento $( A $) cuando puede ocurrir debido a varios **escenarios mutuamente excluyentes y exhaustivos** $( B_1, B_2, \dots, B_n $). Se expresa as√≠:
$
P(A) = \sum_{i=1}^{n} P(B_i) \cdot P(A \mid B_i)
$

Donde:
- $( \{B_1, B_2, \dots, B_n\} $) es una **partici√≥n del espacio muestral** (eventos disjuntos cuya uni√≥n cubre todo).
- $( P(A \mid B_i) $) es la probabilidad condicional de $( A $) dado que ocurri√≥ $( B_i $).
- $( P(B_i) $) es la probabilidad de que ocurra el escenario $( B_i $).

---

### üß† Ejemplo:

Una f√°brica tiene tres m√°quinas que producen tornillos:
- M√°quina A produce el 30% de los tornillos y tiene 2% de defectuosos.
- M√°quina B produce el 50% y tiene 1% de defectuosos.
- M√°quina C produce el 20% y tiene 3% de defectuosos.

¬øCu√°l es la probabilidad de que un tornillo seleccionado al azar sea defectuoso?
$
P(D) = P(A)P(D \mid A) + P(B)P(D \mid B) + P(C)P(D \mid C)
$
$
P(D) = 0.3(0.02) + 0.5(0.01) + 0.2(0.03) = 0.006 + 0.005 + 0.006 = 0.017
$

Entonces, $( P(D) = 0.017 $), es decir, **1.7% de los tornillos son defectuosos**.

---

### üîÑ Teorema de Bayes

El teorema de Bayes **invierte probabilidades condicionales**, permiti√©ndonos calcular la probabilidad de una **causa** dada una **consecuencia**. Se expresa as√≠:
$
P(B_i \mid A) = \frac{P(B_i) \cdot P(A \mid B_i)}{P(A)}
$

Este teorema es muy √∫til para **revisar nuestras creencias** ante nueva evidencia, y es la base del **enfoque bayesiano de la probabilidad**.

---

### üß™ Ejemplo con el mismo caso:

Si se encuentra un tornillo defectuoso, ¬øcu√°l es la probabilidad de que lo haya producido la M√°quina C?
$
P(C \mid D) = \frac{P(C) \cdot P(D \mid C)}{P(D)} = \frac{0.2 \cdot 0.03}{0.017} \approx 0.3529
$
Entonces, la probabilidad de que **la M√°quina C haya producido el tornillo defectuoso** es aproximadamente **35.3%**.

---

### ‚úÖ Conclusi√≥n

- El **teorema de la probabilidad total** permite descomponer una probabilidad complicada en partes m√°s manejables.
- El **teorema de Bayes** nos ayuda a actualizar nuestras creencias cuando obtenemos nueva informaci√≥n.
Ambos teoremas son pilares en el estudio de la probabilidad, tanto en teor√≠a como en aplicaciones reales como diagn√≥stico m√©dico, inteligencia artificial y toma de decisiones.


## Probabilidad Cl√°sica (Laplaciana)

## Definici√≥n Formal

El modelo cl√°sico, tambi√©n llamado de **Laplace**, define la probabilidad como:

$$
P(A) = \frac{|A|}{|S|}
$$

donde:

- $ S $ es el espacio muestral (conjunto de todos los resultados posibles).
- $ A \subseteq S\ $ es un evento.
- Todos los resultados de $( S )$ son **equiprobables**.

## Teorema Fundamental

Teorema: Para cualquier evento A en un espacio muestral finito S con resultados equiprobables, la probabilidad de A est√° dada por la f√≥rmula cl√°sica.

**Demostraci√≥n:**

Sea S un espacio muestral con n resultados posibles, todos igualmente probables.

Por el principio de indiferencia (Laplace), cada resultado tiene probabilidad $\frac{1}{n}$.

Para un evento A con k resultados favorables, la probabilidad es la suma de las probabilidades de los resultados favorables:

$$
P(A) = \sum_{i=1}^{k} \frac{n_1}{n_k} = \frac{n(S)}{n(A)}
$$

## Axiomas de Kolmog√≥rov

Cualquier funci√≥n de probabilidad $ P: \mathcal{F} \rightarrow [0, 1] $ sobre un espacio muestral $ S $, debe cumplir:

1. **Axioma de No Negatividad**:  
   $$
   \forall A \in \mathcal{F}, \quad P(A) \geq 0
   $$

2. **Axioma de Normalizaci√≥n**:  
   $$
   P(S) = 1
   $$

3. **Aditividad Finitaria** (para eventos disjuntos):  
   $$
   A \cap B = \varnothing \Rightarrow P(A \cup B) = P(A) + P(B)
   $$

## Teoremas Fundamentales

#### Teorema 1: Probabilidad del Complemento

$$
P(A^c) = 1 - P(A)
$$

**Demostraci√≥n:**

Como $ A \cup A^c = S $ y $ A \cap A^c = \varnothing $, se cumple:

$$
P(S) = P(A) + P(A^c) \Rightarrow P(A^c) = 1 - P(A)
$$

#### Teorema 2: Probabilidad de la Uni√≥n

$$
P(A \cup B) = P(A) + P(B) - P(A \cap B)
$$

**Demostraci√≥n:**

La uni√≥n cuenta dos veces la intersecci√≥n. Usamos conteo de cardinalidad:

$$
|A \cup B| = |A| + |B| - |A \cap B| \Rightarrow \frac{|A \cup B|}{|S|} = \frac{|A|}{|S|} + \frac{|B|}{|S|} - \frac{|A \cap B|}{|S|}
$$

## Aplicaci√≥n: Lanzamiento de Dos Dados

Queremos calcular la probabilidad de que la **suma de dos dados sea 7**.


```python
# importamos librerias
import itertools
import matplotlib.pyplot as plt

# Espacio muestral: todos los pares (d1, d2)
espacio = list(itertools.product(range(1, 7), repeat=2))
n_total = len(espacio)

# Evento: suma igual a 7
evento = [p for p in espacio if sum(p) == 7]
n_favorables = len(evento)

# Probabilidad cl√°sica
P = n_favorables / n_total
print(f"Probabilidad de suma 7: {P:.4f}")

# Crear datos para el histograma
suma_frecuencias = [sum(p) for p in espacio]
plt.hist(suma_frecuencias, bins=range(2, 14), align='left', rwidth=0.8, color='yellow', edgecolor='black')
plt.axvline(7, color='red', linestyle='--', label='Suma 7')
plt.title('Distribuci√≥n de Sumas de Dos Dados')
plt.xlabel('Suma')
plt.ylabel('Frecuencia')
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

### Ejemplo com√∫n: Lanzamiento de un dado y obtener un n√∫mero par

```python

import matplotlib.pyplot as plt

# Espacio muestral de un dado
resultados = [1, 2, 3, 4, 5, 6]
n_total = len(resultados)

# Evento: obtener un n√∫mero par
evento = [2, 4, 6]
n_favorables = len(evento)

# Calculamos la probabilidad cl√°sica
probabilidad = n_favorables / n_total

# Visualizaci√≥n
labels = ['Favorables', 'No favorables']
sizes = [n_favorables, n_total - n_favorables]
colors = ['lightgray', 'lightpink']

plt.figure(figsize=(6, 6))
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=90)
plt.title(f'Probabilidad cl√°sica: obtener n√∫mero par = {probabilidad:.2f}')
plt.axis('equal')
plt.show()
```


## Probabilidad Frecuentista

## Fundamento Te√≥rico

La probabilidad frecuentista se basa en la **frecuencia relativa** observada de un evento en m√∫ltiples repeticiones del mismo experimento.

Sea $ A $ un evento de inter√©s, y $ n(A) $ el n√∫mero de veces que ocurre despu√©s de $ n $ repeticiones, entonces:

$$
P(A) \approx \frac{n(A)}{n}
$$

Cuando $ n \to \infty $, esta proporci√≥n converge (por la ley de los grandes n√∫meros):

$$
P(A) = \lim_{n \to \infty} \frac{n(A)}{n}
$$

## Teorema Fundamental (Ley de los Grandes N√∫meros)

Teorema (Ley D√©bil de los Grandes N√∫meros)

Sea $X_1, X_2, \dots$ una secuencia de variables aleatorias i.i.d. con $E[X_i] = \mu$.

Entonces, para cualquier $\epsilon > 0$:
$$
\lim_{n \to \infty} P\left( \left| \frac{1}{n} \sum_{i=1}^n X_i - \mu \right| \geq \epsilon \right) = 0
$$

**Demostraci√≥n:** *(usando la desigualdad de Chebyshev)*

Sea $\overline{X}_n = \frac{1}{n} \sum_{i=1}^n X_i$:
$$
E[\overline{X}_n] = \mu \quad \text{y} \quad \text{Var}(\overline{X}_n) = \frac{\sigma^2}{n}
$$

Por Chebyshev:
$$
P\left( |\overline{X}_n - \mu| \geq \epsilon \right) \leq \frac{\text{Var}(\overline{X}_n)}{\epsilon^2} = \frac{\sigma^2}{n \epsilon^2}
$$

Cuando $n \to \infty$, el lado derecho tiende a 0.

## Interpretaci√≥n

- **No se basa en suposiciones de equiprobabilidad**, sino en la observaci√≥n directa.
- Es el fundamento de la estad√≠stica inferencial y la teor√≠a de la estimaci√≥n.

## Simulaci√≥n: Lanzamiento de una moneda

Vamos a simular $ n = 10{,}000 $ lanzamientos de una moneda justa y observar c√≥mo se estabiliza la frecuencia relativa de ‚Äúsol‚Äù.


```python
# importamos librerias
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(42)
n = 10000
resultados = np.random.choice(['sol', 'aguila'], size=n)
frecuencia_cara = np.cumsum(resultados == 'sol') / np.arange(1, n + 1)

plt.figure(figsize=(10, 5))
plt.plot(frecuencia_cara, label='Frecuencia relativa de "sol"', color='black')
plt.axhline(0.5, color='red', linestyle='--', label='Valor te√≥rico: 0.5')
plt.title('Probabilidad Frecuentista: Convergencia de Frecuencia Relativa')
plt.xlabel('N√∫mero de lanzamientos')
plt.ylabel('Frecuencia relativa')
plt.grid(True)
plt.legend()
plt.show()
```

### Ejemplo: Convergencia de frecuencias relativas

```python
# importamos librerias
import numpy as np
import matplotlib.pyplot as plt

# Simulaci√≥n de lanzamiento de moneda
np.random.seed(42)
n_simulations = 10000
results = np.random.choice([0, 1], size=n_simulations, p=[0.4, 0.6])  # Sesgada P(1)=0.6

# Calculamos frecuencias relativas acumuladas
frec_rel = np.cumsum(results) / np.arange(1, n_simulations+1)

# Gr√°fico
plt.figure(figsize=(10, 6))
plt.plot(frec_rel, label='Frecuencia relativa')
plt.axhline(y=0.6, color='black', linestyle='--', label='Probabilidad te√≥rica')
plt.xlabel('N√∫mero de ensayos')
plt.ylabel('Frecuencia relativa de "1"')
plt.title('Convergencia de la probabilidad frecuentista')
plt.legend()
plt.grid(True)

# Zoom en los primeros 1000 ensayos
plt.figure(figsize=(10, 6))
plt.plot(frec_rel[:1000], label='Frecuencia relativa (primeros 1000 ensayos)')
plt.axhline(y=0.6, color='black', linestyle='--', label='Probabilidad te√≥rica')
plt.xlabel('N√∫mero de ensayos')
plt.ylabel('Frecuencia relativa de "1"')
plt.title('Convergencia inicial de la probabilidad frecuentista')
plt.legend()
plt.grid(True)
plt.show()
```


## Probabilidad Geom√©trica

## Fundamento Te√≥rico

La probabilidad geom√©trica se aplica cuando el espacio muestral es **continuo**, como longitudes, √°reas o vol√∫menes.

$$
P(A) = \frac{\mu(A)}{\mu(S)}
$$

donde $ \mu $ es una **medida** (longitud, √°rea, etc.), y $ A \subseteq S $.

## Teorema Fundamental

Teorema: Sea S un espacio muestral con medida finita no nula, y A un subconjunto medible de S. Entonces la probabilidad geom√©trica de A est√° bien definida y satisface los axiomas de probabilidad.

**Demostraci√≥n:**


1. $P(A) = \frac{\mu(A)}{\mu(S)} \geq 0 \text{, ya que } \mu(A) \geq 0 \text{ y } \mu(S) > 0$

2. $P(S) = \frac{\mu(S)}{\mu(S)} = 1$

3. Para una secuencia de eventos mutuamente excluyentes ${A1,A2,....:}$

$$P\left(\bigcup_{i=1}^\infty A_i\right) = \frac{\mu(S)}{\mu\left(\bigcup_{i=1}^\infty A_i\right)} = \sum_{i=1}^\infty \frac{\mu(S)}{\mu(A_i)} = \sum_{i=1}^\infty P(A_i)
$$

## Problema del Encuentro

> Dos personas acuerdan encontrarse entre las 0 y 60 minutos. Cada una espera 15 minutos. ¬øCu√°l es la probabilidad de que se encuentren?

Definimos:

- $ x $: hora de llegada de la persona A
- $ y $: hora de llegada de la persona B
- Espacio muestral: cuadrado de $ [0, 60] \times [0, 60] $
- Evento favorable: $ |x - y| \leq 15 $

## C√°lculo Anal√≠tico

√Årea total: $ 60 \times 60 = 3600 $

√Årea del evento: $ 3600 - 2 \cdot \frac{1}{2} \cdot 45^2 = 1575 $

$
P(\text{encuentro}) = \frac{1575}{3600} = \frac{7}{16}
$


```python

import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 60, 500)
y1 = x - 15
y2 = x + 15

plt.figure(figsize=(8, 8))
plt.fill_between(x, y1, y2, where=(y1 >= 0) & (y2 <= 60), color='lightgreen', alpha=0.6, label='Zona de encuentro')
plt.plot(x, y1, 'b--', label='y = x - 15')
plt.plot(x, y2, 'y--', label='y = x + 15')
plt.xlim(0, 60)
plt.ylim(0, 60)
plt.xlabel('Hora de llegada de A (min)')
plt.ylabel('Hora de llegada de B (min)')
plt.title('Probabilidad Geom√©trica - Problema del Encuentro')
plt.legend()
plt.grid(True)
plt.gca().set_aspect('equal')
plt.show()
```

### Ejemplo: Problema de la aguja de Buffon

```python
# importamos librerias
import numpy as np
import matplotlib.pyplot as plt

def buffon_experimento(n, l=1.0, d=2.0):
    crosses = 0
    for _ in range(n):
        theta = np.random.uniform(0, np.pi / 2)
        y = np.random.uniform(0, d / 2)
        if y <= (l / 2) * np.sin(theta):
            crosses += 1
    return (2 * l * n) / (crosses * d) if crosses > 0 else 0  # Evitar divisi√≥n por cero

# Estimaci√≥n de œÄ
estimates = [buffon_experimento(1000) for _ in range(20)]

# Gr√°fico de estimaciones
plt.figure(figsize=(10, 5))

# Subgr√°fico 1: Estimaciones de œÄ
plt.subplot(1, 2, 1)
plt.plot(estimates, marker='o')
plt.axhline(np.pi, color='r', linestyle='--', label='œÄ')
plt.title('Estimaci√≥n de œÄ con la aguja de Buffon')
plt.xlabel('Simulaci√≥n')
plt.ylabel('Valor estimado')
plt.legend()
plt.grid(True)

# Subgr√°fico 2: Probabilidad geom√©trica
prob = np.mean([buffon_experimento(1000) for _ in range(100)]) / np.pi  # Probabilidad simulada
prob_teorica = (2 / np.pi)  # Probabilidad te√≥rica
pi_est = np.mean(estimates)  # Promedio estimado de œÄ

plt.subplot(1, 2, 2)
plt.bar(['Simulaci√≥n', 'Te√≥rico'], [prob, prob_teorica], color=['lightgray', 'lightblue'])
plt.ylabel('Probabilidad')
plt.title(f'Probabilidad geom√©trica\nEstimaci√≥n promedio de œÄ: {pi_est:.4f}')

# Mostrar gr√°ficos
plt.tight_layout()
plt.show()
```


## Probabilidad Condicional e Independencia

## Probabilidad Condicional

$$
P(A \mid B) = \frac{P(A \cap B)}{P(B)}, \quad \text{si } P(B) > 0
$$

### Interpretaci√≥n:

Es la probabilidad de que ocurra $ A $ **sabemos que** ocurri√≥ $ B $. Se reduce el espacio muestral a $ B $.

## Independencia de Eventos

Dos eventos $ A $ y $ B $ son **independientes** si:

$$
P(A \cap B) = P(A) \cdot P(B) \quad \text{o equivalentemente} \quad P(A \mid B) = P(A)
$$

## Ejemplo: Dos dados

- $ A $: suma de los dados = 7
- $ B $: primer dado impar


```python
# importamos librerias y paquetes
import itertools
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Espacio muestral
S = list(itertools.product(range(1, 7), repeat=2))
total = len(S)

# Eventos
A = [p for p in S if sum(p) == 7]
B = [p for p in S if p[0] % 2 == 1]
A_inter_B = [p for p in A if p in B]

# C√°lculos
P_A = len(A) / total
P_B = len(B) / total
P_A_inter_B = len(A_inter_B) / total
P_A_dado_B = P_A_inter_B / P_B
independencia = np.isclose(P_A, P_A_dado_B)

print(f"P(A) = {P_A:.4f}")
print(f"P(B) = {P_B:.4f}")
print(f"P(A ‚à© B) = {P_A_inter_B:.4f}")
print(f"P(A | B) = {P_A_dado_B:.4f}")
print(f"¬øSon independientes? {'S√≠' if independencia else 'No'}")
```


## Teoremas de la Probabilidad Total y de Bayes

## Teorema de la Probabilidad Total

Si $ A_1, A_2, \dots, A_n $ son una partici√≥n de $ S $, y $ B $ es cualquier evento, entonces:

$$
P(B) = \sum_{i=1}^n P(A_i) \cdot P(B \mid A_i)
$$

## Teorema de Bayes

$$
P(A_k \mid B) = \frac{P(A_k) \cdot P(B \mid A_k)}{\sum_{i=1}^n P(A_i) \cdot P(B \mid A_i)}
$$

## Ejemplo: Clasificaci√≥n de correos electr√≥nicos

Supongamos que:
 - El 20% de los correos son spam (P(Spam) = 0.2).
 - El 80% de los correos no son spam (P(NoSpam) = 0.8).
 - Si un correo es spam, hay un 90% de probabilidad de que contenga ciertas palabras clave (P(ClaveSpam) = 0.9).
 - Si un correo no es spam, hay un 10% de probabilidad de que contenga esas palabras clave (P(ClaveNoSpam) = 0.1).

Queremos calcular:
- La probabilidad total de que un correo contenga las palabras clave (P(Clave)).
- La probabilidad de que un correo sea spam si contiene las palabras clave (P(SpamClave)), usando el Teorema de Bayes.



```python
# Probabilidades iniciales
P_Spam = 0.2  # Probabilidad de que un correo sea spam
P_NoSpam = 1 - P_Spam  # Probabilidad de que un correo no sea spam
P_Clave_Spam = 0.9  # Probabilidad de clave dado que es spam
P_Clave_NoSpam = 0.1  # Probabilidad de clave dado que no es spam

# Probabilidad total de que el correo contenga las palabras clave (P(Clave))
P_Clave = P_Spam * P_Clave_Spam + P_NoSpam * P_Clave_NoSpam

# Probabilidad de que un correo sea spam dado que contiene las palabras clave (P(Spam|Clave))
P_Spam_dado_Clave = (P_Spam * P_Clave_Spam) / P_Clave

# Mostrar los resultados
print(f"Probabilidad total de que un correo contenga las palabras clave (P(Clave)) = {P_Clave:.4f}")
print(f"Probabilidad de que un correo sea spam dado que contiene las palabras clave (P(Spam|Clave)) = {P_Spam_dado_Clave:.4f}")

# Visualizaci√≥n gr√°fica
import matplotlib.pyplot as plt

# Datos para la gr√°fica
etiquetas = ["P(Clave)", "P(Spam|Clave)"]
valores = [P_Clave, P_Spam_dado_Clave]

# Crear la gr√°fica
plt.figure(figsize=(8, 6))
plt.bar(etiquetas, valores, color=['lightblue', 'blue'], edgecolor='black')
plt.title('Probabilidades en Clasificaci√≥n de Spam', fontsize=16)
plt.ylabel('Probabilidad', fontsize=14)
plt.ylim(0, 1)  # Escala de 0 a 1
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Mostrar la gr√°fica
plt.tight_layout()
plt.show()
```
