## I. Estimadores

En estad√≠stica inferencial, un **estimador** es una funci√≥n de los datos muestrales que se utiliza para aproximar un **par√°metro poblacional desconocido**, como la media $\mu$, la varianza $\sigma^2$, o la proporci√≥n $p$.

Si se quiere estimar un par√°metro $\theta$ de la poblaci√≥n, se busca una funci√≥n $\hat{\theta}$ (basada en la muestra) que lo aproxime lo mejor posible. Esta funci√≥n se llama **estimador** y su valor num√©rico calculado a partir de los datos se llama **estimaci√≥n**.

---

### üéØ Ejemplos comunes de estimadores

- Estimador de la media poblacional $\mu$:
  $$
  \hat{\mu} = \overline{X} = \frac{1}{n} \sum_{i=1}^{n} X_i
  $$

- Estimador de la varianza poblacional $\sigma^2$:
  $$
  \hat{\sigma}^2 = \frac{1}{n-1} \sum_{i=1}^{n} (X_i - \overline{X})^2
  $$

- Estimador de la proporci√≥n poblacional $p$:
  $$
  \hat{p} = \frac{x}{n}
  $$
  donde $x$ es el n√∫mero de √©xitos en una muestra de tama√±o $n$.

---

### üß† Propiedades deseables de un estimador

Un buen estimador debe tener ciertas propiedades estad√≠sticas. Las m√°s importantes son:

---

### ‚úÖ Insesgadez (o ausencia de sesgo)

Un estimador $\hat{\theta}$ es **insesgado** si su esperanza matem√°tica es igual al par√°metro que estima:

$$
\mathbb{E}[\hat{\theta}] = \theta
$$

Esto significa que, en promedio, el estimador "acierta".

---

### üìâ Varianza y eficiencia

De entre todos los estimadores insesgados, se prefiere el que tenga **menor varianza**, ya que tiene menor dispersi√≥n en torno al valor verdadero.

- Un estimador con varianza peque√±a es **m√°s preciso**.
- Si un estimador tiene la menor varianza posible dentro de una clase de estimadores, se dice que es **eficiente**.

---

### üìâ Consistencia

Un estimador $\hat{\theta}_n$ es **consistente** si, a medida que el tama√±o muestral $n$ crece, el estimador se aproxima al valor real del par√°metro:

$$
\hat{\theta}_n \xrightarrow{P} \theta \quad \text{cuando } n \to \infty
$$

Esto significa que, con suficientes datos, el estimador converge (en probabilidad) al par√°metro.

---

### üßÆ Suficiencia

Un estimador es **suficiente** si contiene toda la informaci√≥n disponible en la muestra sobre el par√°metro. Es decir, no hay p√©rdida de informaci√≥n al usar ese estimador en lugar de los datos originales.

---

### ‚úèÔ∏è Recapitulaci√≥n

Las principales propiedades que buscamos en un estimador son:

- **Insesgadez** ‚Üí que sea correcto "en promedio".
- **Varianza baja / eficiencia** ‚Üí que tenga poca dispersi√≥n.
- **Consistencia** ‚Üí que se acerque al valor real con m√°s datos.
- **Suficiencia** ‚Üí que aproveche toda la informaci√≥n de la muestra.

Estas propiedades ayudan a comparar distintos estimadores y decidir cu√°l usar en cada caso.

---

### ‚úÖ Conclusi√≥n

Los estimadores son herramientas clave en estad√≠stica inferencial. Sus propiedades nos permiten evaluar qu√© tan confiables son nuestras conclusiones cuando trabajamos con muestras. Elegir un buen estimador es esencial para tomar decisiones basadas en datos.

## II. Intervalos de Confianza

En estad√≠stica inferencial, los **intervalos de confianza** son herramientas fundamentales para **estimar par√°metros poblacionales** a partir de datos muestrales. En lugar de ofrecer un solo valor como estimaci√≥n (estimador puntual), se proporciona un **rango de valores posibles** que, con cierta confianza, contiene al verdadero par√°metro.

---

### üß† ¬øQu√© es un intervalo de confianza?

Un **intervalo de confianza (IC)** para un par√°metro $\theta$ es un intervalo calculado a partir de los datos de una muestra, del tipo:

$$
[\text{L√≠mite\ inferior}, \text{L√≠mite\ superior}]
$$

Este intervalo est√° asociado a un nivel de confianza $1 - \alpha$, usualmente expresado como porcentaje (por ejemplo, 95%).

Interpretaci√≥n com√∫n:  
> Si repiti√©ramos el muestreo muchas veces, aproximadamente el 95% de los intervalos construidos contendr√≠an el verdadero valor del par√°metro.

---

### üßÆ Estructura general de un IC

Un intervalo de confianza suele tener la forma:

$$
\hat{\theta} \pm \text{margen de error}
$$

Donde:
- $\hat{\theta}$: estimador puntual del par√°metro $\theta$.
- Margen de error = valor cr√≠tico √ó error est√°ndar.

---

### üìå Intervalo de confianza para la media (cuando $\sigma$ es conocida)

Si $X_1, \dots, X_n$ son i.i.d. con media $\mu$ y desviaci√≥n est√°ndar conocida $\sigma$, entonces el intervalo de confianza del 95% para $\mu$ es:

$$
\overline{X} \pm z_{\alpha/2} \cdot \frac{\sigma}{\sqrt{n}}
$$

Donde:
- $\overline{X}$: media muestral.
- $z_{\alpha/2}$: valor cr√≠tico de la distribuci√≥n normal est√°ndar. Para 95%, $z_{0.025} \approx 1.96$.

---

### üìå Intervalo de confianza para la media (cuando $\sigma$ es desconocida)

En este caso se usa la **distribuci√≥n t de Student** con $n - 1$ grados de libertad:

$$
\overline{X} \pm t_{\alpha/2, n-1} \cdot \frac{S}{\sqrt{n}}
$$

Donde $S$ es la desviaci√≥n est√°ndar muestral.

---

### üìå Intervalo de confianza para una proporci√≥n

Si $x$ es el n√∫mero de √©xitos en una muestra de tama√±o $n$, el IC para la proporci√≥n $p$ es:

$$
\hat{p} \pm z_{\alpha/2} \cdot \sqrt{ \frac{\hat{p}(1 - \hat{p})}{n} }
$$

Donde $\hat{p} = x / n$ es la proporci√≥n muestral.

---

### üìè Nivel de confianza

El nivel de confianza m√°s com√∫n es 95%, pero tambi√©n se usan 90% y 99%. Un nivel de confianza m√°s alto da un **intervalo m√°s amplio**, y uno m√°s bajo da un **intervalo m√°s estrecho**.

---

### ‚ö†Ô∏è ¬°Ojo con la interpretaci√≥n!

Un intervalo de confianza del 95% **no significa** que hay 95% de probabilidad de que el par√°metro est√© en ese intervalo. El par√°metro es fijo; lo que var√≠a es el intervalo.

La interpretaci√≥n correcta es:
> En el 95% de las muestras posibles, el intervalo construido incluir√° el verdadero par√°metro poblacional.

---

### ‚úÖ Conclusi√≥n

Los intervalos de confianza nos permiten cuantificar la incertidumbre de nuestras estimaciones. Son herramientas poderosas y ampliamente utilizadas en ciencia, econom√≠a, ingenier√≠a y muchas otras disciplinas donde se toman decisiones con base en datos.

## III. Pruebas de Hip√≥tesis

Las **pruebas de hip√≥tesis** son procedimientos estad√≠sticos que nos permiten tomar decisiones o extraer conclusiones sobre una poblaci√≥n, a partir de los resultados obtenidos de una muestra.

Este proceso se basa en la idea de **contrastar una afirmaci√≥n (hip√≥tesis)** sobre un par√°metro poblacional con la informaci√≥n que proviene de los datos.

---

### üéØ Objetivo

Determinar si hay suficiente evidencia en los datos muestrales para **rechazar una afirmaci√≥n previa (hip√≥tesis nula)** sobre la poblaci√≥n.

---

### üß© Hip√≥tesis Nula y Alternativa

- **Hip√≥tesis nula ($H_0$)**: Es la afirmaci√≥n inicial que se pone a prueba. Normalmente representa una **posici√≥n conservadora o sin efecto**.
- **Hip√≥tesis alternativa ($H_1$ o $H_a$)**: Es la afirmaci√≥n que se acepta si se rechaza $H_0$. Representa una diferencia o efecto significativo.

Ejemplo:
$$
H_0: \mu = 100 \quad \text{vs} \quad H_1: \mu \ne 100
$$

---

### üî¢ Tipos de prueba

Dependiendo de la forma de $H_1$, existen tres tipos de prueba:

1. **Bilateral**:
   $$
   H_1: \theta \ne \theta_0
   $$
2. **Unilateral a la derecha**:
   $$
   H_1: \theta > \theta_0
   $$
3. **Unilateral a la izquierda**:
   $$
   H_1: \theta < \theta_0
   $$

---

### ‚öôÔ∏è Procedimiento general

1. **Plantear las hip√≥tesis** ($H_0$ y $H_1$).
2. **Elegir el nivel de significancia** $\alpha$ (por ejemplo, 0.05).
3. **Seleccionar la estad√≠stica de prueba** adecuada.
4. **Calcular el valor de la estad√≠stica** con los datos muestrales.
5. **Determinar la regi√≥n cr√≠tica o calcular el p-valor**.
6. **Tomar una decisi√≥n**:
   - Si el valor cae en la regi√≥n cr√≠tica o $p$-valor $< \alpha$, se **rechaza $H_0$**.
   - Si no, se **falla al rechazar $H_0$**.

---

### üß™ Estad√≠sticas de prueba comunes

- Para la media (con $\sigma$ conocida):
  $$
  Z = \frac{\overline{X} - \mu_0}{\sigma / \sqrt{n}}
  $$

- Para la media (con $\sigma$ desconocida):
  $$
  t = \frac{\overline{X} - \mu_0}{S / \sqrt{n}}
  $$

- Para proporciones:
  $$
  Z = \frac{\hat{p} - p_0}{\sqrt{ \frac{p_0(1 - p_0)}{n} }}
  $$

---

### üìâ Error tipo I y II

- **Error tipo I**: Rechazar $H_0$ cuando es verdadera. Su probabilidad es $\alpha$.
- **Error tipo II**: No rechazar $H_0$ cuando es falsa. Su probabilidad es $\beta$.

La **potencia de una prueba** es $1 - \beta$ y representa la capacidad de detectar un efecto cuando realmente existe.

---

### üìä El p-valor

El **p-valor** es la probabilidad, bajo $H_0$, de obtener un resultado tan extremo como el observado. Si el p-valor es menor que $\alpha$, hay evidencia suficiente para rechazar $H_0$.

---

### ‚úÖ Conclusi√≥n

Las pruebas de hip√≥tesis son herramientas esenciales para la toma de decisiones basada en evidencia. Permiten aceptar o rechazar afirmaciones sobre par√°metros poblacionales, bajo un marco riguroso y cuantificable. Es crucial plantear bien las hip√≥tesis, entender el contexto y elegir adecuadamente la prueba estad√≠stica.

### Estimadores

### Definici√≥n formal
Un estimador es una funci√≥n de los datos muestrales que aproxima un par√°metro poblacional desconocido. Matem√°ticamente, si $ X_1, X_2, \dots, X_n $ son variables aleatorias que representan una muestra, un estimador $ \hat{\theta} $ del par√°metro $ \theta $ es una funci√≥n:

$$
\hat{\theta} = g(X_1, X_2, \dots, X_n)
$$

###  Propiedades de los estimadores

#### Insesgadez
Un estimador $ \hat{\theta} $ es insesgado si su esperanza matem√°tica coincide con el par√°metro poblacional:

$$
E[\hat{\theta}] = \theta
$$

**Ejemplo:**
- La media muestral $ \bar{X} $ es un estimador insesgado de $ \mu $, ya que $ E[\bar{X}] = \mu $.

- La varianza muestral $ S^2 = \frac{1}{n-1} \sum (X_i - \bar{X})^2 $ es insesgada para $ \sigma^2 $, mientras que $ \frac{1}{n} \sum (X_i - \bar{X})^2 $ es sesgada.

#### Consistencia
Un estimador es consistente si converge en probabilidad al par√°metro verdadero cuando $ n \to \infty $:

$$
\lim_{n \to \infty} P(|\hat{\theta} - \theta| > \epsilon) = 0 \quad \forall \epsilon > 0
$$

**Condiciones suficientes:**
- Insesgadez asint√≥tica $$ \lim_{n \to \infty} E[\hat{\theta}] = \theta $$.
- Varianza decreciente $$ \lim_{n \to \infty} \text{Var}(\hat{\theta}) = 0 $$.

####  Eficiencia
Dos estimadores insesgados $ \hat{\theta}_1 $ y $ \hat{\theta}_2 $ se comparan mediante su varianza. El m√°s eficiente es el de menor varianza:

$$
\text{Var}(\hat{\theta}_1) < \text{Var}(\hat{\theta}_2) \implies \hat{\theta}_1 \text{ es m√°s eficiente.}
$$

**Ejemplo:**
- La media muestral es el estimador m√°s eficiente de \( \mu \) bajo distribuci√≥n normal.

#### Suficiencia
Un estimador es suficiente si contiene toda la informaci√≥n relevante sobre $ \theta $ presente en la muestra.

**Criterio de factorizaci√≥n de Fisher-Neyman:**

$$
f(X_1, \dots, X_n \mid \theta) = g(\hat{\theta} \mid \theta) \cdot h(X_1, \dots, X_n)
$$

donde $ h $ no depende de $ \theta $.

---

### Intervalos de confianza (IC)


### Concepto y fundamentaci√≥n
Un intervalo de confianza al $ (1-\alpha) \times 100\% $ para un par√°metro $ \theta $ es un rango $ (L, U) $ tal que:

$$
P(L \leq \theta \leq U) = 1 - \alpha
$$

**Interpretaci√≥n correcta:** Si construy√©ramos muchos ICs, el $ (1-\alpha)\% $ de ellos contendr√≠a el verdadero $ \theta $.

### IC para la media poblacional $ \mu $

**Caso 1:** Varianza poblacional $ \sigma^2 $ conocida:

$$
IC = \bar{X} \pm z_{\alpha/2} \cdot \frac{\sigma}{\sqrt{n}}
$$

**Ejemplo:** Ilustrativo con c√≥digo Python para calcular intervalos de confianza.

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

# Datos
muestra = np.array([10.2, 9.8, 10.5, 10.1, 9.9])
sigma = 0.5  # conocido
nivel_confianza = 0.95

# C√°lculo
media = np.mean(muestra)
z_critico = stats.norm.ppf(1 - (1 - nivel_confianza) / 2)
margen_error = z_critico * (sigma / np.sqrt(len(muestra)))
ic = (media - margen_error, media + margen_error)

print(f"IC del 95% para Œº: ({ic[0]:.3f}, {ic[1]:.3f})")

# Graficar los resultados
x = np.linspace(media - 3*sigma, media + 3*sigma, 100)
y = stats.norm.pdf(x, loc=media, scale=sigma / np.sqrt(len(muestra)))

plt.figure(figsize=(8, 5))
plt.plot(x, y, label="Distribuci√≥n Normal", color="blue")
plt.axvline(ic[0], color="red", linestyle="--", label=f"L√≠mite inferior: {ic[0]:.3f}")
plt.axvline(ic[1], color="green", linestyle="--", label=f"L√≠mite superior: {ic[1]:.3f}")
plt.axvline(media, color="orange", linestyle="-", label=f"Media: {media:.3f}")
plt.title(f"Intervalo de confianza al {nivel_confianza*100:.0f}% para Œº")
plt.xlabel("Valor")
plt.ylabel("Densidad de probabilidad")
plt.legend()
plt.grid()
plt.show()
```

**Caso 2:**  Varianza poblacional desconocida (usando t de Student)

Cuando la varianza poblacional $( \sigma^2 $) es desconocida, se utiliza la distribuci√≥n t de Student para construir el intervalo de confianza.

#### F√≥rmula del Intervalo de Confianza:
$$
IC = \bar{X} \pm t_{\alpha/2, n-1} \cdot \frac{S}{\sqrt{n}}
$$

Donde:
- $ \bar{X} $: Media muestral.
- $ t_{\alpha/2, n-1} $: Valor cr√≠tico de la distribuci√≥n t de Student con $ n-1 $ grados de libertad.
- $ S $: Desviaci√≥n est√°ndar muestral.
- $ n $: Tama√±o de la muestra.

#### Interpretaci√≥n:
El intervalo de confianza $ IC $ proporciona un rango en el cual es probable que se encuentre el verdadero valor de la media poblacional $ \mu $, con un nivel de confianza de $ 1-\alpha $.



```python
# importamos librerias
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

# Datos de ejemplo
np.random.seed(42)  # Para reproducibilidad
n = 30  # Tama√±o de la muestra
mu = 50  # Media real de la poblaci√≥n
sigma = 10  # Desviaci√≥n est√°ndar real de la poblaci√≥n

# Generar muestra aleatoria
sample = np.random.normal(loc=mu, scale=sigma, size=n)
sample_mean = np.mean(sample)
sample_std = np.std(sample, ddof=1)  # Desviaci√≥n est√°ndar muestral

# Nivel de confianza
alpha = 0.05  # Nivel de significancia
confidence = 1 - alpha
df = n - 1  # Grados de libertad
t_critical = stats.t.ppf(1 - alpha / 2, df)  # Valor cr√≠tico t

# Intervalo de confianza
margin_of_error = t_critical * (sample_std / np.sqrt(n))
ci_lower = sample_mean - margin_of_error
ci_upper = sample_mean + margin_of_error

print(f"Media muestral: {sample_mean:.2f}")
print(f"Intervalo de confianza al {confidence*100:.0f}%: ({ci_lower:.2f}, {ci_upper:.2f})")

# Graficar la distribuci√≥n y el intervalo de confianza
x = np.linspace(sample_mean - 4*sample_std, sample_mean + 4*sample_std, 100)
y = stats.t.pdf((x - sample_mean) / (sample_std / np.sqrt(n)), df)

plt.figure(figsize=(8, 5))
plt.plot(x, y, label="Distribuci√≥n t de Student", color="blue")
plt.axvline(ci_lower, color="red", linestyle="--", label=f"L√≠mite inferior: {ci_lower:.2f}")
plt.axvline(ci_upper, color="green", linestyle="--", label=f"L√≠mite superior: {ci_upper:.2f}")
plt.axvline(sample_mean, color="orange", linestyle="-", label=f"Media muestral: {sample_mean:.2f}")
plt.title(f"Intervalo de confianza al {confidence*100:.0f}% para la media")
plt.xlabel("Valor")
plt.ylabel("Densidad de probabilidad")
plt.legend()
plt.grid()
plt.show()
```

## Intervalos de Confianza para la proporci√≥n poblacional $p$


La f√≥rmula es:

$$
IC = \hat{p} \pm z_{\alpha / 2} \cdot \sqrt{\frac{\hat{p}(1 - \hat{p})}{n}}
$$

Descripci√≥n del c√≥digo

- Entradas: Se define la proporci√≥n muestral $(\hat{p} = 0.6)$, el tama√±o de muestra $(n = 100)$, y el nivel de significancia $(\alpha = 0.05)$.

- C√°lculos: Se obtienen el valor cr√≠tico $z_{\alpha/2}$, el error est√°ndar, y los l√≠mites inferior y superior del intervalo de confianza.

- Gr√°fica: Se visualiza la distribuci√≥n normal de la proporci√≥n muestral, destacando el intervalo de confianza y la proporci√≥n muestral.

```python
# importamos librerias
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Par√°metros iniciales
p_hat = 0.6  # Proporci√≥n muestral
n = 100      # Tama√±o de la muestra
alpha = 0.05 # Nivel de significancia (para un nivel de confianza del 95%)

# C√°lculo del valor cr√≠tico z_alpha/2
z_alpha_over_2 = norm.ppf(1 - alpha / 2)

# C√°lculo del error est√°ndar
error_estandar = np.sqrt((p_hat * (1 - p_hat)) / n)

# C√°lculo del intervalo de confianza
lim_inferior = p_hat - z_alpha_over_2 * error_estandar
lim_superior = p_hat + z_alpha_over_2 * error_estandar

# Resultados
print(f"Intervalo de confianza del 95%: [{lim_inferior:.4f}, {lim_superior:.4f}]")

# Gr√°fica del intervalo de confianza
x = np.linspace(0, 1, 500)
y = norm.pdf(x, loc=p_hat, scale=error_estandar)

plt.figure(figsize=(8, 5))
plt.plot(x, y, label="Distribuci√≥n de la proporci√≥n muestral", color="blue")
plt.axvline(lim_inferior, color='red', linestyle='--', label=f"L√≠mite inferior: {lim_inferior:.4f}")
plt.axvline(lim_superior, color='green', linestyle='--', label=f"L√≠mite superior: {lim_superior:.4f}")
plt.axvline(p_hat, color='orange', linestyle='-', label=f"Proporci√≥n muestral: {p_hat:.4f}")

plt.title("Intervalo de confianza para la proporci√≥n poblacional")
plt.xlabel("Proporci√≥n poblacional")
plt.ylabel("Densidad de probabilidad")
plt.legend()
plt.grid()
plt.show()
```

### Pruebas de hip√≥tesis

Las pruebas de hip√≥tesis son procedimientos para evaluar afirmaciones sobre par√°metros poblacionales.

### Pasos b√°sicos:
1. Establecer hip√≥tesis nula $(H_0$) y alternativa $(H_1$).
2. Seleccionar nivel de significancia $(\alpha$).
3. Calcular estad√≠stico de prueba.
4. Determinar regi√≥n cr√≠tica o valor $p$.
5. Tomar decisi√≥n rechazar o no rechazar $(H_0$).

### F√≥rmulas comunes:
- Prueba $z$ para media $(\sigma)$ conocida:

  $$
  z = \frac{\bar{x} - \mu_0}{\sigma / \sqrt{n}}
  $$

- Prueba $t$ para media $(\sigma)$ desconocida:
  $$
  t = \frac{\bar{x} - \mu_0}{s / \sqrt{n}}
  $$

- Prueba $z$ para proporciones:
  $$
  z = \frac{\hat{p} - p_0}{\sqrt{p_0 (1 - p_0) / n}}
  $$

## Prueba de hip√≥tesis para la media ($ \sigma $ desconocida)

### Hip√≥tesis
- **Hip√≥tesis nula ($ H_0 $):** $ \mu = \mu_0 $
- **Hip√≥tesis alternativa $( H_1 )$:**
  - $ \mu \neq \mu_0 $ (prueba bilateral)
  - $ \mu > \mu_0 $ (prueba unilateral derecha)
  - $ \mu < \mu_0 $ (prueba unilateral izquierda)

### Estad√≠stico de prueba
$$
t = \frac{\bar{x} - \mu_0}{s / \sqrt{n}}
$$

### Regla de decisi√≥n
- **Prueba bilateral:** Rechazar $ H_0 $ si $ |t| > t_{\alpha/2, n-1} $
- **Prueba unilateral derecha:** Rechazar $ H_0 $ si $ t > t_{\alpha, n-1} $
- **Prueba unilateral izquierda:** Rechazar $ H_0 $ si $ t < -t_{\alpha, n-1} $

### Alternativamente:
Rechazar $ H_0 $ si el valor $p$ $ < \alpha $.

**Ejemplo:**

```python
# importamos librerias
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt

# Datos y par√°metros de prueba
muestra = [28.5, 29.3, 28.7, 29.0, 29.5, 28.9, 29.2, 28.8]
mu0 = 28.0  # Valor hipot√©tico
alpha = 0.05

# Calcular estad√≠stico de prueba
x_barra = np.mean(muestra)
s = np.std(muestra, ddof=1)
n = len(muestra)
t_stat = (x_barra - mu0) / (s / np.sqrt(n))

# Valor cr√≠tico (prueba bilateral)
t_critico = stats.t.ppf(1 - alpha / 2, df=n-1)

# Valor p
valor_p = 2 * (1 - stats.t.cdf(abs(t_stat), df=n-1))

print(f"Estad√≠stico t: {t_stat:.4f}")
print(f"Valor cr√≠tico: ¬±{t_critico:.4f}")
print(f"Valor p: {valor_p:.4f}")

# Decisi√≥n
if abs(t_stat) > t_critico:
    print("Rechazar H‚ÇÄ: hay evidencia suficiente para concluir que Œº ‚â† 28.0")
else:
    print("No rechazar H‚ÇÄ: no hay evidencia suficiente para concluir que Œº ‚â† 28.0")

# Graficar la distribuci√≥n t y los resultados
x = np.linspace(-4, 4, 500)  # Eje x para la curva t de Student
y = stats.t.pdf(x, df=n-1)  # Densidad de probabilidad t

plt.figure(figsize=(10, 6))
plt.plot(x, y, label="Distribuci√≥n t de Student", color="blue")
plt.fill_between(x, y, where=(x < -t_critico) | (x > t_critico), color="red", alpha=0.3, label="Regi√≥n de rechazo")
plt.axvline(t_stat, color="orange", linestyle="--", label=f"Estad√≠stico t: {t_stat:.4f}")
plt.axvline(-t_critico, color="green", linestyle="--", label=f"Valor cr√≠tico: -{t_critico:.4f}")
plt.axvline(t_critico, color="green", linestyle="--", label=f"Valor cr√≠tico: {t_critico:.4f}")
plt.title("Prueba de hip√≥tesis: comparaci√≥n del estad√≠stico t y la regi√≥n cr√≠tica")
plt.xlabel("Valores de t")
plt.ylabel("Densidad de probabilidad")
plt.legend()
plt.grid()
plt.show()
```
